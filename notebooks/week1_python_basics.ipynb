```python
import numpy as np
import pandas as pd
```

```python
arr = np.array([10, 20, 30, 40])
print("Array:", arr)
print("Mean:", arr.mean())
print("Standard Deviation:", arr.std())
```

```python
df = pd.DataFrame({"Values": arr})
df.describe()
```

```python
salaries = pd.DataFrame({
    "Employee": ["A", "B", "C", "D"],
    "Salary": [50000, 60000, 75000, 80000]
})
salaries
```

```python
salaries.loc[2, "Salary"] = None
salaries["Salary"].fillna(salaries["Salary"].median(), inplace=True)
salaries
```
